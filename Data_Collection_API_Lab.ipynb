```python
import requests
import pandas as pd
import numpy as np

```

```python
# API endpoint for the SpaceX launch data
spacex_url = "https://api.spacexdata.com/v4/launches"

# Get data from the API
response = requests.get(spacex_url)

# Check response
print(response.status_code)

# Convert response to JSON
data = response.json()

# Convert JSON to DataFrame
data_falcon9 = pd.json_normalize(data)

# Display first 5 rows
data_falcon9.head()
```

```python
# Calculate the mean of PayloadMass column
payload_mean = data_falcon9['payloads'].apply(lambda x: len(x) if isinstance(x, list) else np.nan).mean()

# Replace missing values (just an example, modify based on actual column)
data_falcon9['PayloadMass'] = data_falcon9['payloads'].apply(lambda x: len(x) if isinstance(x, list) else payload_mean)

# Save to CSV
data_falcon9.to_csv('dataset_part_1.csv', index=False)
```

```python
columns = ['name', 'date_utc', 'rocket', 'payloads', 'launchpad', 'cores', 'success', 'details']
data_falcon9 = data_falcon9[columns]
data_falcon9.head()
```
